```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
print("Hello Jupyter Notebook")
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head(5)
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.scatter(df['year'], df['per capita income (US$)'], color='red', marker='+')
```

```python
reg_model = linear_model.LinearRegression()
reg_model.fit(df[['year']], df['per capita income (US$)'])
```

```python
reg_model.predict([[2020]]) #2D array, x_samples, x_features
```

```python
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.scatter(df['year'], df['per capita income (US$)'], color='red', marker='+')
plt.scatter(df['year'], reg_model.predict(df[['year']]), color='blue')
```

```python
to_predict_df = pd.read_csv("to_predict.csv")
to_predict_df.head(5)
```

```python
income_vals = reg_model.predict(to_predict_df) #2D array, x_samples, x_features
```

```python
(to_predict_df["Per Capita Income (US$)"]) = np.ceil(income_vals)
to_predict_df.head(10)
```

```python
to_predict_df.to_csv("predicted_income.csv", index=False)
```
